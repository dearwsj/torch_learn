### Tensor
Tensor is a fundamental data structure in PyTorch, essentially a multi-dimensional array. It can be a scalar, a 1D array (vector), a 2D array (matrix), or a higher-dimensional array. Tensors are similar to NumPy ndarrays, but with the added benefit of GPU acceleration. Their API closely resembles NumPy and MATLAB. Let's walk through some examples of basic tensor operations.

```python
import torch as t

print("PyTorch version: ", t.__version__)
```

If you have an NVIDIA GPU, you can leverage CUDA (Compute Unified Device Architecture) to accelerate training on the graphics card.

```python
cuda_available = t.cuda.is_available()

print(f"CUDA availability status: {cuda_available}")
print(f"CUDA version: {t.version.cuda}")
```

```python
if cuda_available:
    print(f"Current graphics card: {t.cuda.get_device_name(0)}")
    
    x = t.rand(5, 3).cuda()

    
    print("Successfully created a tensor on the GPU:")
    print(x)
    
else:
    print("No GPU detected. Please check your installation steps.")

# device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# x = torch.rand(5, 3).to(device)
```

```python
# Construct a 5x3 matrix with allocated memory but uninitialized values
x = t.Tensor(5, 3)

x = t.Tensor([[1,2],[3,4]])
x
```

```python
# Initialize a 2D array with random values from a uniform distribution over [0, 1]
x = t.rand(5,3)
print(x)

# Initialize a 2D array with random values from a standard normal distribution
y = t.randn(5,3)
print(y)
```

```python
print(x.size()) # Check the shape of x
print(x.size()[1]) # Get the number of columns — two equivalent ways
print(x.size(1))
```

```python
x + y # Addition: method 1
```

```python
t.add(x, y) # Addition: method 2
```

```python
# Addition: method 3
result = t.Tensor(5, 3) # Pre-allocate memory
t.add(x, y, out=result) # Output the result
result
```

```python
print('Initial y')
print(y)

print('After add(), y is:')
y.add(x)  # Regular addition, y remains unchanged
print(y)

print('After add_(), y is:')
y.add_(x)  # In-place addition, y is modified
print(y)
```

```python
print(x[:, 1]) # torch.Size([5])     — 1D
print(x[:, 1:2]) # torch.Size([5, 1])  — 2D
```

Tensors support a wide range of operations, including mathematical computations, linear algebra, indexing, and slicing, with an API very similar to NumPy.

Converting between Tensors and NumPy arrays is seamless and efficient. For operations not directly supported by Tensors, you can convert to a NumPy array, perform the operation, and convert back.

```python
a = t.ones(5)
a
```

```python
b = a.numpy() # Tensor -> Numpy
b
```

```python
import numpy as np
a = np.zeros(5)
b = t.from_numpy(a) # Numpy -> Tensor
print(a)
print(b)
```

Tensors and NumPy arrays share the same underlying memory, so conversion between them is fast and virtually cost-free. However, this also means that modifying one will affect the other.

```python
b.add_(1)  # Functions ending with `_` modify the tensor in-place
print(a)
print(b)  # Tensor and NumPy array share the same memory
```

```python
scalar_b0 = b[0]
scalar_b0
```

```python
scalar_b0.size() # 0-dim tensor
```

```python
# .item() works on any tensor with a single element, regardless of its dimensions. It will raise an error if the tensor contains more than one element
scalar_b0.item()
```

```python
tensor_1 =  t.tensor([1]) # tensor_1 is 1-dimensional, scalar_b0 is 0-dimensional
tensor_1, scalar_b0
```

```python
tensor_1.size(), scalar_b0.size()
```

```python
tensor_1.item()
```

Additionally, PyTorch provides `torch.tensor`, an interface very similar to `np.array`. The two are used in much the same way. Note the distinction from `torch.Tensor`, which is a class.

```python
tensor = t.tensor([3,4]) # Create a new tensor containing the elements 3 and 4
tensor
```

```python
scalar = t.tensor(3)
scalar
```

```python
old_tensor = tensor
new_tensor = old_tensor.clone()
new_tensor[0] = 1111
old_tensor, new_tensor
```

Note that `t.tensor()` and `tensor.clone()` always copy the data, so the new tensor no longer shares memory with the original. If you want to share memory, use `torch.from_numpy()` or `tensor.detach()` instead — both create a new tensor that shares the underlying memory.

`.detach()` separates a tensor from the computation graph and returns a new tensor that shares the same underlying memory but no longer participates in gradient computation.

```python
new_tensor = old_tensor.detach()
new_tensor[0] = 1111
old_tensor, new_tensor
```

A Tensor can be moved to the GPU using the `.cuda()` method to take advantage of GPU-accelerated computation.

```python
# On machines without CUDA support, the next step will still run on the CPU
# You only need to specify the device index when you have multiple GPUs — for example, "cuda:1" refers to the second GPU. With a single GPU, "cuda" and "cuda:0" are equivalent.
device = t.device("cuda" if t.cuda.is_available() else "cpu")

x = x.to(device)
y = y.to(x.device)
z = x @ y
z
```

Additionally, you can use `tensor.cuda()` to copy a tensor to the GPU, but this approach is not recommended.

Transferring data from system memory to GPU memory incurs additional overhead. The advantages of GPU computation only become apparent with large-scale data and complex operations.
